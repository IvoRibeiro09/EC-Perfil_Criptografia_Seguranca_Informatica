**2. Implemente um protótipo do esquema descrito  na norma FIPS 205 que deriva do algoritmo SPHINCS+.**

```python
from sphincs import Sphincs
```

Implementação do "SLH-DSA-SHAKE-256s" consideramos os seguintes parametros:

* n = 32;
* h = 64;
* d = 8;
* k = 22;
* a = 14;
* m = 47;

Com ojetivo de:

* pk-bytes = 64;
* sign-bytes = 29792;

```python
sphincs = Sphincs()
sk, pk = sphincs.spx_keygen()
print(f"Secret-Key:\n{sk}\nlen: {len(sk)}")
print(f"Public-Key:\n{pk}\nlen: {len(pk)}")
```

```python
message = b'secret message to send!'
sign = sphincs.spx_sign(message, sk)
print(f"Sign:\n{sign}\nlen: {len(sign)}")
```

```python
validation = sphincs.spx_verify(message, sign, pk)
print("Validation:", validation)
```
